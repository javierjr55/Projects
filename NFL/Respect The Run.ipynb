We begin by importing the proper packages and also bringing in the data.

```python
#import packages
import pandas as pd
import numpy as np
# from sklearn.model_selection import train_test_split
from sklearn import metrics as mt
from sklearn import preprocessing
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings("ignore")

#import data set into data frame under alias NFL
ORIG_NFL = pd.read_csv("NFL.csv")

#Extract the following variables to make a new dataframe to use for the model
nfl = ORIG_NFL[['two_point_attempt','posteam_score', 'defteam_score', 'score_differential', 'posteam_timeouts_remaining', 
                'defteam_timeouts_remaining', 'shotgun', 'no_huddle', 'play_type', 'ydstogo', 'down', 'drive', 'qtr', 
                'game_seconds_remaining', 'yardline_100', 'posteam_type']].copy()
```

The `play_type` class has the following levels:
>run - the ball carrier run with the ball <br>
pass - the quarterback throws the ball to a receiver <br>
field_goal - the kicker attempts to kick the ball through the uprights <br>
qb_spike - the quarterback spikes the ball in order to stop the game clock <br>
qb_kneel - the quarterback takes a knee in order to keep the clock running <br>
punt - the punter punts the ball over to the opposing team in an exchange of possession <br>
extra_point - the kicker attempts to kick the ball through the uprights after scoring a touchdown <br>
kickoff - the scoring team kicks the ball to the opposing team in an exchange of possession <br>
N/A - a play is not called as a result of an infraction, timeout, or clock expiration <br>

In football, a strong running game results in control of possession, which can ultimately lead to control of the game tempo. For this very reason, defenses know that they must <i>always</i> respect the run! As such, we will focus on trying to predict when an offense is likely to run the ball. In order to do so, we bring in only the `play_type` scenarios in which a run is possible and start to clean up the data set.

```python
#View the variables to determine if Pandas read them into the dataframe accordingly
nfl = nfl[(nfl.play_type == 'run') | (nfl.play_type == 'pass') | (nfl.play_type == 'field_goal') | 
          (nfl.play_type == 'qb_spike') | (nfl.play_type == 'qb_kneel') | (nfl.play_type == 'punt')]
nfl = nfl.fillna(0)

#We see that several variables were read in as floats when they should be integers.
#List of the variables misclassified as floats
cols = ['two_point_attempt', 'posteam_score', 'defteam_score', 'score_differential', 'posteam_timeouts_remaining',
        'defteam_timeouts_remaining', 'down', 'game_seconds_remaining', 'yardline_100']

#Convert variable list to integers
nfl[cols] = nfl[cols].applymap(np.int64)

#verify conversion took place
nfl.info()
```

We can see there are 14 int(64) variables and 2 objects in the data set. The `posteam_type` variable looks at whether the team with possession of the ball is either the <i>home</i> or <i>away</i> team. We can transform this variable to `home_team_pos` to simply determine if the team with possession is the home team.<br><br>

We will also transpose the `play_type` into running plays only, which will equal 1 and all other to equal 0.

```python
#transpose play_types into different columns and add back into dataframe
tmp_nfl = pd.get_dummies(nfl.play_type,prefix='play_type')
nfl = pd.concat((nfl,tmp_nfl),axis=1)

#create new variable in which posteam_type now refers to home_possession
nfl['home_team_pos'] = nfl.posteam_type == 'home' 
nfl.home_team_pos = nfl.home_team_pos.astype(np.int64) #convert to int64

#drop all excess columns
nfl = nfl.drop(['posteam_type','play_type','play_type_field_goal', 'play_type_pass', 'play_type_qb_kneel', 'play_type_qb_spike', 'play_type_punt'], axis = 1)

nfl.play_type_run = nfl.play_type_run.astype(np.int64) #convert to int64

#create a copy for future analysis
nfl1 = nfl.copy()

#view current dataframe to ensure conversions occurred
nfl.info()
```

### Test & Train Split
With the dataset now ready, we begin with the cross validation preparation. We create a test set based on 20% of the data and a train set of 80% of the data.

```python
from sklearn.model_selection import ShuffleSplit

#creating the X and y
if 'play_type_run' in nfl:\
y = nfl['play_type_run'].values # get the labels we want
del nfl['play_type_run'] # get rid of the class label
X = nfl.values # use other columns to predict

num_cv_iterations = 10 #iteration to perform cross-validations
num_instances = len(y) #size of data frame

#create the split
cv_object = ShuffleSplit(n_splits=num_cv_iterations,
                         test_size  = 0.2)
                         
print(cv_object)    
```

### Logistic Regression Model
In this instance we test the model using different cost values (0.001, 1.0, 2.5, 5.0, 10) in order to determine which model would produce the highest accuracy. Aside from accuracy, we are also looking at execution time of the model. Each model will run 10 times and then display the average accuracy from those 10 iterations. Based on the average accuracy and performance, we can then decide which cost value would be better suited for the logistic regression moving forward.

```python
%%time # function to provide execution time

# logistic model
lr_clf = LogisticRegression(penalty='l2', C=0.001, class_weight=None)

# create empty variables for loop
iter_num=0
t_acc=0
# loop for logistic regression model
for train_indices, test_indices in cv_object.split(X,y): 

    X_train = X[train_indices]
    y_train = y[train_indices]
    
    X_test = X[test_indices]
    y_test = y[test_indices]

    lr_clf.fit(X_train,y_train)  # train model
    y_hat = lr_clf.predict(X_test) # test model

    acc = mt.accuracy_score(y_test,y_hat) # accuracy of model for given instance
    t_acc = t_acc+acc # running total of accuracies
    iter_num+=1 # iteration counter
    
print('Avg. Accuracy (cost = 0.001):',t_acc/10*100,'%') # print average accuracy of model
```

Based on the tests above, we can see a cost value of 10 produced slightly higher accuracy than the other models. The performance was also only slightly better than the rest. With the desired model in place for logistic regression, we now look to the features in the model. However, in order to do so we must normalize the feautures accordingly.

```python
from sklearn.preprocessing import StandardScaler
weights = lr_clf.coef_.T # take transpose to make a column vector of coefficients
variable_names = nfl.columns # get column names

# scale attributes by the training set
scl_obj = StandardScaler()
scl_obj.fit(X_train) # fit scaler on train set

X_train_scaled = scl_obj.transform(X_train) # apply to training
X_test_scaled = scl_obj.transform(X_test) # apply those means and std to the test set (without snooping at the test set values)

# train the model just as before
lr_clf = LogisticRegression(penalty='l2', C=10.0) # model
lr_clf.fit(X_train_scaled,y_train)  # train

y_hat = lr_clf.predict(X_test_scaled) # get test set precitions

acc = mt.accuracy_score(y_test,y_hat) # accuracy
conf = mt.confusion_matrix(y_test,y_hat) # confusion matrix
print('accuracy:', acc*100,'%' )
print(conf )

# sort these attributes and print
zip_vars = zip(lr_clf.coef_.T,nfl.columns)
zip_vars = sorted(zip_vars)
for coef, name in zip_vars:
    print(name, 'has weight of', coef[0])
```

Based on the current logistic regression, we now have the weights of the variables normalized and ready to interpret. However, the model is not as efficient as it could be. Because of the method in which logistic regression executes the model, the current execution time is roughly 1 minute. Considering we have a large data set (over 300,000 instances), we can optimize the model much more using stochastic gradient descent.

```python
from sklearn.model_selection import StratifiedShuffleSplit 
from sklearn.linear_model import SGDClassifier

# test and train for stochastic gradient descent classifier
# use 50/50 split this time
cv = StratifiedShuffleSplit( n_splits=1,test_size=0.5)

# create variables for model below
regularize_const = 0.1 # alpha
iterations = 50 # number of times to perform model
```

Unlike before where we were splitting the training and test set on an 80/20, we will opt for a 50/50 split this time around. However, we will now perform the model a total of 50 times unlike the 10 times from before. 

```python
%%time
# logistic regression classifier with stochastic descent
log_sgd = SGDClassifier(alpha=regularize_const,
        fit_intercept=True, l1_ratio=0.0, learning_rate='optimal',
        loss='log', n_iter=iterations, n_jobs=-1, penalty='l2')

scl = StandardScaler()

#loop for model to perform n iterations
for train_idx, test_idx in cv.split(X,y):
    log_sgd.fit(scl.fit_transform(X[train_idx]),y[train_idx])
    yhat = log_sgd.predict(scl.transform(X[test_idx]))
    
    conf = mt.confusion_matrix(y[test_idx],yhat)
    acc = mt.accuracy_score(y[test_idx],yhat)

print('Logistic Regression:', acc)
```

Using logistic regression with stochastic gradient descent, we were able to optimze our execution time down to 5 seconds! Although we lost a 1% accuracy, the gain in execution is well worth the sacrifice. However, we are not done just yet. We can continue the optimization by now reducing the demensionality of the set. In this instance, we will accept 5 components in the reduced model and see how well it performs compared to the newly optimized logistic regression with stochastic gradient descent.

```python
%%time
# now lets use some of what we know from this class to reduce the dimensionality of the set
from sklearn.decomposition import PCA
n_components = 5

pca = PCA(n_components=n_components, svd_solver='randomized')

log_sgd = SGDClassifier(alpha=regularize_const,
        fit_intercept=True, l1_ratio=0.0, learning_rate='optimal',
        loss='log', n_iter=iterations, n_jobs=-1, penalty='l2')

# you could also set this up in a pipeline
for train_idx, test_idx in cv.split(X,y):
    log_sgd.fit(pca.fit_transform(X[train_idx]),y[train_idx])
    yhat = log_sgd.predict(pca.transform(X[test_idx]))
    
    conf = mt.confusion_matrix(y[test_idx],yhat)
    acc = mt.accuracy_score(y[test_idx],yhat)

print('Logistic Regression:', acc)
```

We can see that execution time remained about the same and we now have a large sacrifice in accuracy in exchange for a reduction in dimension. In this instance, it would probably be wise to opt for the prior model. Because the variables are all given prior to snap of the ball, we can afford to keep all the variables in the logistic regression model.

### Support Vector Machine
With the Logistic Regression model now complete and optimized, we turn over to the Support Vector Machine model. Considering we have a considerably large data set, we opted for the Linear Support Vector Machine model using stochastic gradient descent. Just like the previous model, we begin by splitting the data into a test and train set and scaling it for the SVM.

```python
# create the training and testing set
for train_indices, test_indices in cv_object.split(X,y): 

    X_train = X[train_indices]
    y_train = y[train_indices]
    
    X_test = X[test_indices]
    y_test = y[test_indices]
    
X_train_scaled = scl_obj.transform(X_train) # apply to training
X_test_scaled = scl_obj.transform(X_test) # apply to test
```

With the test and training sets in place, we now begin to apply them to the linear support vector machine model. Because of how the SVM executes the model, we are only performing one iteration of the model. The model runs multiple vectors through it in order to fit itself to the support vectors. 

```python
%%time
from sklearn.svm import LinearSVC

svm_clf = LinearSVC(penalty='l2', C=5.0) # create model
svm_clf.fit(X_train_scaled, y_train) #fit model

y_hat = svm_clf.predict(X_test_scaled) # test model


acc = mt.accuracy_score(y_test,y_hat)
conf = mt.confusion_matrix(y_test,y_hat)
print('Accuracy:', acc*100,'%' )
print('==Conf. Matrix==')
print(conf)
print('================')
```

Using a cost value of 5.0, we were able to get considerably close to the accuracy of the logistic regression model! However, the execution time is more than double than that of the logistic regression model. However, just like with the logistic regression model, we can also apply stochastic gradient descent to the model in order to attempt to optimize the performance.

```python
%%time

# create the classifier
svm_sgd = SGDClassifier(alpha=regularize_const,
        fit_intercept=True, l1_ratio=0.0, learning_rate='optimal',
        loss='hinge', n_iter=iterations, n_jobs=-1, penalty='l2')

scl = StandardScaler()

#run the model using the classifier
for train_idx, test_idx in cv.split(X,y):
    svm_sgd.fit(scl.fit_transform(X[train_idx]),y[train_idx])
    yhat = svm_sgd.predict(scl.transform(X[test_idx]))
    
    conf = mt.confusion_matrix(y[test_idx],yhat)
    acc = mt.accuracy_score(y[test_idx],yhat)

print('SVM:', acc)
```

Using the SGD Classifier, we were able to optimize our Linear SVM model exceptionally well in terms of performance. The performance of the model has now been reduced to just under 4 seconds with virtually the same accuracy as before!
