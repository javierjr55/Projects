For this project, we will be using this data set to determine if the `play_type` is going to be a <i>run</i> play.
<br><br>
With over 350,000 records, this data set is an ideal candidate for the `StratifiedShuffleSplit` method. <i>"This cross-validation object is a merge of StratifiedKFold and ShuffleSplit, which returns stratified randomized folds. The folds are made by preserving the percentage of samples for each class"</i> [1](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html). By opting for this route, we are able to get a blend of cross-validation while preserving computing power. Unlike other methods such as `StratifiedKFold` and `ShuffleSplit`, the `StratifiedShuffleSplit` works well with large data sets like this one and harnesses the effectiveness of other methods.
<br><br>
We will use precision as our performance metric. The cost of a false-negative is much higher than the cost of a false-positive. While accurancy is a strong metric within its own right, its limitations don't allow it to penalize the errors as much as it should. Failure to recognize that significance could be the difference between a scoring play or a turnover, which could change the course of the game's final outcome. For this very reason, precision will help us find the right model while also ensuring we account for the true cost of the errors.

<br><br>The table above shows the desired variables and their data type we would need in order to be able to run our models for the prediction. The `play_type` variable would get transposed to only account for running plays as events. This new variable gets named `play_type_run` in the table. The `posteam_type` would also get transposed to only count home teams as events, which is why it is then renamed `home_team_pos` in the table. However, before we get to the final dataset we desire we need to check for any missing/null values in the data set.

```python
#Extract the following variables to make a new dataframe to use for the model
nfl_run = ORIG_NFL[['two_point_attempt', 'score_differential', 'posteam_timeouts_remaining', 
                'defteam_timeouts_remaining', 'shotgun', 'no_huddle', 'play_type', 'ydstogo', 'down', 'drive', 'qtr', 
                'game_seconds_remaining', 'yardline_100', 'posteam_type', 'desc']].copy()

#View the variables to determine if Pandas read them into the dataframe accordingly
nfl_run = nfl_run[(nfl_run.play_type == 'run') | (nfl_run.play_type == 'pass') | (nfl_run.play_type == 'field_goal') | 
          (nfl_run.play_type == 'qb_spike') | (nfl_run.play_type == 'qb_kneel') | (nfl_run.play_type == 'punt')]

# Check for NaN values throughout the data set
nfl_run.isnull().sum(axis = 0)
```

Based on the results from the data set, it appears we have some NULL values. First, we take a look at the `score_differential` variable to determine whether the NaN are justified or missing data.

```python
# Show NaN values for posteam_score
nfl_run[np.isnan(nfl_run.score_differential)]
```

Looking into the data set, it appears the NaN values for `score_differential` are all kickoff plays, which are clearly identified by looking at the time, quarter, and drive of the play. Considering here is no potential for the offense to execute a run play from this play attempt, we are safe to remove these plays from the data set.

```python
# Delete all null values for posteam_score
nfl_run = nfl_run[pd.notnull(nfl_run['score_differential'])]

# Show null values for game_seconds_remaining
nfl_run[np.isnan(nfl_run.game_seconds_remaining)]
```

Interestingly enough, there doesn't appear to be any observable reason as to why the `game_seconds_remaining` value is null in these instances. The game clock in the play description indicates that the plays all occurred in the 5:00, 3:20, or 1:40 mark. Considering these instances appear to be errors and only 18 instances in a data set with nearly 400,000 instances, we will go ahead and remove them from the data set.
<Br><br>
Next, we look towards `down`, which appears to have over 700 NaNs in the data set.

```python
# Delete all null values for game_seconds_remaining
nfl_run = nfl_run[pd.notnull(nfl_run['game_seconds_remaining'])]

# Since there are over 700 instances of null down values, we'll only look at a few
# in an effort to gain insight into why the nulls are appearing.
temp_df = nfl_run[np.isnan(nfl_run.down)]
temp_df.head()
```

The `down` variable has NaN values only attributed to the two-point attempts. This makes sense considering the two-point attempts are not part of the chain of downs. They are the equivalent of a 4th down in the sense that it is the final attempt before possession goes to the other team (unless the 4th down is converted into a 1st down). However, this play is not considered a down play and therefore should not have a down (as the NaN would indicate). As a result, we will label all NaN values in the `down` variable to 0. By doing this, we should have now addressed all of the NaN values in the data set and are ready to proceed to transposing the variables.

```python
# Convert all remaining NaN values into 0
nfl_run = nfl_run.fillna(0)

# Verify no more nulls
nfl_run.isnull().sum(axis = 0)
```

By doing this, we should have now addressed all of the NaN values in the data set and are ready to proceed to transposing the variables. We create the `play_type_run` variable to only identify run plays in the data set. We also create the `home_team_pos` variable to identify when the home team has possession of the ball. Ultimately, we drop the variables we don't need and ensure all the remaining variables are integers.

```python
#Several variables were read in as floats when they should be integers.
#List of the variables misclassified as floats
cols = ['two_point_attempt', 'score_differential', 'posteam_timeouts_remaining',
        'defteam_timeouts_remaining', 'down', 'game_seconds_remaining', 'yardline_100']

#Convert variable list to integers
nfl_run[cols] = nfl_run[cols].applymap(np.int64)

#transpose play_types into different columns and add back into dataframe
tmp_nfl_run = pd.get_dummies(nfl_run.play_type,prefix='play_type')
nfl_run = pd.concat((nfl_run,tmp_nfl_run),axis=1)

#create new variable in which posteam_type now refers to home_possession
nfl_run['home_team_pos'] = nfl_run.posteam_type == 'home' 
nfl_run.home_team_pos = nfl_run.home_team_pos.astype(np.int64) #convert to int64

#drop all excess columns
nfl_run = nfl_run.drop(['posteam_type','play_type','play_type_field_goal', 'play_type_pass', 'play_type_qb_kneel', 'play_type_qb_spike', 'play_type_punt', 'desc'], axis = 1)

nfl_run.play_type_run = nfl_run.play_type_run.astype(np.int64) #convert to int64

nfl_run.info()
```

With the data types finalized and the missing values addressed, we now look towards the data set. Considering there are other play types besides run, we anticipated the distribution to be lopsided. While our assumptions were correct, we prepared for such a distribution by adjusting our validation analysis accordingly. In the plot below, we can see there appear to be nearly 100,000 more instances in which the play was other than a run,

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use("ggplot")
sns.countplot(x='play_type_run', data=nfl_run, palette='hls')
plt.title('Play Type = Run')  
plt.xlabel('Play Type')
plt.ylabel('Events')
plt.show
```

While we know the plays are in favor of all other, we can also see a skewed distribution in the `yardline_100` variable. Once again, this is to be expected considering most offenses start their offenses at the 20 yardline. Furthermore, as the rules change then so will the concetration of plays. As a result, we won't give it too much imprtance for the time being.

### `nfl_run` Data Split
As previously mentioned, I will be using the `StratifiedShuffleSplit` method to split the train and test data set.

```python
from sklearn.model_selection import StratifiedShuffleSplit 

#create checkpoint
nfl_run_copy = nfl_run.copy()

#creating the X and y
if 'play_type_run' in nfl_run:\
y_run = nfl_run['play_type_run'].values # get the labels we want
del nfl_run['play_type_run'] # get rid of the class label
X_run = nfl_run.values # use other columns to predict

#create the split
cv_object = StratifiedShuffleSplit( n_splits=10,test_size=0.2)
print(cv_object)
```

### `RandomForestClassifier` without Cluster Analysis
In order to measure the effect clustering has on our model's performance, we will go ahead and run the model before and after clustering. Once complete, we will be able to look back and analyze the change in performance.

```python
%%time
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# create forest
rf_clf = RandomForestClassifier(max_depth=150, n_estimators=150, n_jobs=-1, oob_score=True)

# run model and get accuracy and precision
acc = cross_val_score(rf_clf,X_run,y=y_run,cv=cv_object)
prec = cross_val_score(rf_clf,X_run,y=y_run,cv=cv_object, scoring='precision')

# print results
print ("Average accuracy =", acc.mean()*100, "+/-", acc.std()*100)
print ("Average precision =", prec.mean()*100, "+/-", prec.std()*100)
```
